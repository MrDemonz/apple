```python
import torch
from PIL import Image
from torchvision import transforms
import matplotlib.pyplot as plt

# Load and rename checkpoint keys
ckpt = torch.load('/content/model200_dark_faces.pth', map_location='cpu')
new_state_dict = {k.replace('conv', 'e_conv'): v for k, v in ckpt.items()}

# Instantiate model and load weights
model = enhance_net_nopool()
model.load_state_dict(new_state_dict)
model.eval()

# Load your image and preprocess
image_path = '/content/0045.png'
image = Image.open(image_path).convert('RGB')
transform = transforms.Compose([
    transforms.ToTensor(),
])
input_img = transform(image).unsqueeze(0)  # Add batch dim

# Run model inference (without GPU)
with torch.no_grad():
    enhance_image_1, enhance_image, r = model(input_img)

# Convert tensor back to image for visualization
def tensor_to_image(tensor):
    tensor = tensor.squeeze(0).clamp(0,1)  # remove batch dim & clamp
    tensor = tensor.permute(1, 2, 0).cpu().numpy()  # CHW to HWC
    tensor = (tensor * 255).astype('uint8')
    return Image.fromarray(tensor)

# Show original and enhanced image
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(image)

plt.subplot(1,2,2)
plt.title("Enhanced")
plt.imshow(tensor_to_image(enhance_image_1))
plt.show()
```

```python
import numpy as np
from PIL import Image
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

# Helper to convert tensor to image
def tensor_to_image(tensor):
    tensor = tensor.squeeze(0).clamp(0, 1)  # remove batch dim & clamp
    tensor = tensor.permute(1, 2, 0).cpu().numpy()  # CHW to HWC
    tensor = (tensor * 255).astype('uint8')
    return Image.fromarray(tensor)

# Load original image (again, if needed)
original = Image.open('/content/0045.png').convert('RGB')

# Convert both to same size if needed
enhanced_image = tensor_to_image(enhance_image_1)
original_resized = original.resize(enhanced_image.size)

# Convert to NumPy arrays in [0, 1] range
original_np = np.array(original_resized).astype(np.float32) / 255.0
enhanced_np = np.array(enhanced_image).astype(np.float32) / 255.0

# Compute PSNR and SSIM
psnr_val = psnr(original_np, enhanced_np, data_range=1.0)
ssim_val = ssim(original_np, enhanced_np, data_range=1.0, channel_axis=-1)

# Output
print(f"PSNR: {psnr_val:.2f} dB")
print(f"SSIM: {ssim_val:.4f}")
```

```python
from PIL import Image
import torch
import numpy as np  # ✅ This line is necessary
from realesrgan import RealESRGANer
from basicsr.archs.rrdbnet_arch import RRDBNet

# Step 1: Define the model architecture
model = RRDBNet(
    num_in_ch=3, num_out_ch=3, num_feat=64,
    num_block=23, num_grow_ch=32, scale=4
)

# Step 2: Initialize RealESRGAN
upscaler = RealESRGANer(
    scale=4,
    model_path='/content/RealESRGAN_x4plus.pth',
    model=model,
    tile=0,
    tile_pad=10,
    pre_pad=0,
    half=False,
    device='cpu'  # or 'cuda' if available
)

# Step 3: Load the image
img = Image.open('/content/0045.png').convert('RGB')

# Step 4: Convert PIL to NumPy
img_np = np.array(img)

# Step 5: Enhance the image
output, _ = upscaler.enhance(img_np, outscale=4)

# Step 6: Show or save the output
output_image = Image.fromarray(output)
output_image.show()
```
